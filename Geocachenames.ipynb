```python
from geonamescache import GeonamesCache
gc = GeonamesCache()
```

```python
# Get the continents
continents = gc.get_continents()
print(continents.keys())
```

```python
south_america = continents['SA']
print(south_america)
```

```python
for name_key in ['name', 'asciiName', 'toponymName']:
    print(south_america[name_key])
```

```python
countries = gc.get_countries()
num_countries = len(countries)
print(f'GeonameCache holds data for {num_countries} countries')

brazil_data = countries['BR']
print('The following data pertains to Brazil')
print(brazil_data)
```

```python
brazil_neighbors = brazil_data['neighbours']
for neighbor_code in brazil_neighbors.split(','):
    print(countries[neighbor_code]['name'])
```

```python
# Get countries by name
result = gc.get_countries_by_names()['Brazil']
assert result == countries['BR']
```

```python
# Accessing City Information
cities = gc.get_cities()
num_cities = len(cities)
print(f"GeoNamesCache holds data for {num_cities} total cities")

city_id = list(cities.keys())[0]
print(cities[city_id])

print(cities)
```

```python
# Fetching Brazil cities from GeoNamesCache
us_cities = [city for city in cities.values()
             if city['countrycode'] == 'US']
num_us_cities = len(us_cities)

print(f"GeoNamesCache holds data for {num_us_cities} Brazil cities.")
print(us_cities)
```

```python
# Fetching cities by name
matched_cities_by_name = gc.get_cities_by_name('Belo Horizonte')
print(matched_cities_by_name)
```

```python
# Fetching multiple cities with a shared name
city_name = 'Louisiana'
matched_cities_list = gc.get_cities_by_name(city_name)

for i, city in enumerate(matched_cities_list):
    city_info = list(city.values())[0]
    country_code = city_info['countrycode']
    country = countries[country_code]['name']
    print(f"The {city_name} at index {i} is located in country {country}")
```

```python
from unidecode import unidecode
accented_names = [city['name'] for city in gc.get_cities().values()
                  if city['name'] != unidecode(city['name'])]
num_accented_cities = len(accented_names)

print(f"An example accented city name is '{accented_names[0]}'")
print(f"{num_accented_cities} cities have accented names")
```

```python
alternative_names = {unidecode(name): name
                     for name in accented_names}
print(gc.get_cities_by_name(alternative_names['Hagatna']))
```
